```python
# import panda package
import pandas as pd

# read dataset and assign it to a Pandas Data Frame
df = pd.DataFrame()
hearts_df = pd.read_csv('/Users/hiroko/Desktop/CS4661(Data)/homework/HW3/data/Heart_s.csv')
```

#### b-

```python
# checking dataset by every 10 lines:
hearts_df[0::10]
```

#### c-

```python
# build feature matrix (ignore categorical features)
feature_cols = ['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak']

X = hearts_df[feature_cols]  

# print the first 5 rows
X.head()

y = hearts_df['AHD']

y[::10]
```

#### d-

```python
from sklearn.model_selection import train_test_split

# split the data set into train and testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

#### e-

```python
from sklearn.metrics import accuracy_score

print('Accuracy of 3 Classifiers \n')

# KNN (with k = 5)
from sklearn.neighbors import KNeighborsClassifier
k = 5
knn = KNeighborsClassifier(n_neighbors=k) 
knn.fit(X_train, y_train)
knn_y_predict = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_y_predict)

print('KNN:', knn_accuracy)

# Decision Tree
from sklearn.tree import DecisionTreeClassifier
my_decisiontree = DecisionTreeClassifier()
my_decisiontree.fit(X_train, y_train)
dicisitontree_y_predict = my_decisiontree.predict(X_test)
dicisiontree_accuracy = accuracy_score(y_test, dicisitontree_y_predict)

print('Decision Tree: ', dicisiontree_accuracy)

# Logistic Regression
from sklearn.linear_model import LogisticRegression
my_logreg = LogisticRegression()
my_logreg.fit(X_train, y_train)
y_predict_lr = my_logreg.predict(X_test)
log_accuracy = accuracy_score(y_test, y_predict_lr)


print('Logistic Regression: ', log_accuracy)
```

Therefore, the best classifiers is logistic regression, and the worst is Decision Tree. 

#### f-

```python
# One Hot Encoding


# Gender (f/m)
def male_to_numeric(x):
    if x == 'm':
        return 1
    elif x == 'f':
        return 0

def female_to_numeric(x):
    if x == 'f':
        return 1
    elif x == 'm':
        return 0
    

# ChestPain (typical/ asymptomatic /nonanginal/ nontypical)
def typical_to_numeric(x):
    if x == 'typical':
        return 1
    else:
        return 0

    
def asymptomatic_to_numeric(x):
    if x == 'asymptomatic':
        return 1
    else:
        return 0

def nonanginal_to_numeric(x):
    if x == 'nonanginal':
        return 1
    else:
        return 0

def nontypical_to_numeric(x):
    if x == 'nontypical':
        return 1
    else:
        return 0


# Thal (fixed / normal / reversable)
def fixed_to_numeric(x):
    if x == 'fixed':
        return 1
    else:
        return 0

def normal_to_numeric(x):
    if x == 'normal':
        return 1
    else:
        return 0

def reversable_to_numeric(x):
    if x == 'reversable':
        return 1
    else:
        return 0
    

# Gender (f/m)
hearts_df['male'] = hearts_df['Gender'].apply(male_to_numeric)
hearts_df['female'] = hearts_df['Gender'].apply(female_to_numeric)


# ChestPain (typical/ asymptomatic /nonanginal/ nontypical)
hearts_df['typical'] = hearts_df['ChestPain'].apply(typical_to_numeric)
hearts_df['asymptomatic'] = hearts_df['ChestPain'].apply(asymptomatic_to_numeric)
hearts_df['nonanginal'] = hearts_df['ChestPain'].apply(nonanginal_to_numeric)
hearts_df['nontypical'] = hearts_df['ChestPain'].apply(nontypical_to_numeric)

# Thal (fixed / normal / reversable)
hearts_df['fixed'] = hearts_df['Thal'].apply(fixed_to_numeric)
hearts_df['normal'] = hearts_df['Thal'].apply(normal_to_numeric)
hearts_df['reversable'] = hearts_df['Thal'].apply(reversable_to_numeric)

# checking dataset by every 10 lines:
hearts_df[0::10]

```

#### g-

```python
# build feature matrix 
feature_cols = ['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak','male','female',
                'typical','asymptomatic','nonanginal','nontypical',
                'fixed','normal','reversable']

X = hearts_df[feature_cols]  

# print the first 5 rows
X.head()

y = hearts_df['AHD']

y[::10]
```

```python
# split the data set into train and testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
print('Accuracy of 3 Classifiers after Hot Encoding \n')

# KNN (with k = 5)
from sklearn.neighbors import KNeighborsClassifier
k = 5
knn = KNeighborsClassifier(n_neighbors=k) 
knn.fit(X_train, y_train)

knn_y_predict = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_y_predict)
print('KNN:', knn_accuracy)

# Decision Tree
from sklearn.tree import DecisionTreeClassifier
my_decisiontree = DecisionTreeClassifier()
my_decisiontree.fit(X_train, y_train)
dicisitontree_y_predict = my_decisiontree.predict(X_test)
dicisiontree_accuracy = accuracy_score(y_test, dicisitontree_y_predict)
print('Decision Tree: ', dicisiontree_accuracy)

# Logistic Regression
from sklearn.linear_model import LogisticRegression
my_logreg = LogisticRegression()
my_logreg.fit(X_train, y_train)
y_predict_lr = my_logreg.predict(X_test)
log_accuracy = accuracy_score(y_test, y_predict_lr)

print('Logistic Regression: ', log_accuracy)

```

#### Accuracy of 3 Classifiers AFTER Hot Encoding

Comparing to two results(before hot encoding and after hot encoding),
